# Experiment Registry Demo

This notebook loads an experiment registry from a CSV/Excel file and uses `load_experiment_dataset` to create `Dataset` objects for selected experiments.

```python
# 1. Set up environment and imports
import os

import pandas as pd

from mfethuls.experiments import load_experiment_registry
from mfethuls.config_loader import load_experiment_dataset

# Point to the CSV/Excel registry maintained by experimentalists.
# For now we use the template CSV shipped in the repo.
EXP_REGISTRY_PATH = os.environ.get('PATH_TO_REGISTRY')

print("Using registry:", EXP_REGISTRY_PATH)

df_registry = load_experiment_registry(EXP_REGISTRY_PATH)
df_registry
```

```python
# 2. Select experiments to analyse

# Example: select only experiments marked as 'to_analyse'
if "status" in df_registry.columns:
    to_analyse = df_registry[df_registry["status"] == "to_analyse"]["name"].tolist()
else:
    to_analyse = df_registry["name"].tolist()

print("Selected experiments:", to_analyse)
to_analyse
```
